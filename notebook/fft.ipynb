# Demo Frequency domain with FFT

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d
from skimage.metrics import structural_similarity as ssim
import os
```

### Find frequency spectrum

```python
image_Path = r'D:\py\Digital_Image_Processing\Image\Image_for_Frequency\demo_fft.jpg' 
I =  cv2.imread(image_Path, cv2.IMREAD_GRAYSCALE)

# Compute 2D FFT
fft_image = np.fft.fft2(I)

# Shift zero frequency to center
fft_shifted = np.fft.fftshift(fft_image)

# Compute magnitude spectrum (amplitude spectrum)
magnitude_spectrum = np.abs(fft_shifted)

# Convert to log scale for better visualization
log_magnitude = np.log(1 + magnitude_spectrum)

# Display original image and amplitude spectrum
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(I, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(log_magnitude, cmap='gray')
plt.title("Amplitude Spectrum")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.plot(magnitude_spectrum)
plt.title("Amplitude Spectrum")
```

### Find output image with Ideal LPF

```python
image_Path = r'D:\py\Digital_Image_Processing\Image\hinhlaco.jpg'
I = cv2.imread(image_Path, cv2.IMREAD_GRAYSCALE)

SI = np.fft.fft2(I)
SI_shift = np.fft.fftshift(SI)

SI_magnitude = np.abs(SI_shift)
SI_log_magnitude = np.log2(SI_magnitude)

H = np.zeros_like(I, dtype=np.float64)
h, w = H.shape

D0 = 20

center_y, center_x = h // 2, w // 2

for i in range(h):
    for j in range(w):
        D = np.sqrt((i - center_y)**2 + (j - center_x)**2)
        if D <= D0:
            H[i, j] = 1
            
        else:
            H[i, j] = 0

SY = SI_shift*H

Y = np.fft.ifft2(SY)

Y = np.abs(Y).astype(np.uint8)

SY_magnitude = np.abs(SY)
SY_log_magnitude = np.log2(SY_magnitude)

plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(I, cmap="gray")
plt.title('Original image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(SI_log_magnitude, cmap="gray")
plt.title('Original image spectrum')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(H, cmap="gray")
plt.title('Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(SY_log_magnitude, cmap="gray")
plt.title('After filter image spectrum')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(Y, cmap="gray")
plt.title('After filter')
plt.axis('off')

plt.tight_layout()
```

### Find output with Butterworth LFP

```python
H_butterworth = np.zeros_like(I, dtype=np.float64)

n = 2

for i in range(h):
    for j in range(w):
        D = np.sqrt((i - center_y)**2 + (j - center_x)**2)
        H_butterworth[i, j] = 1 / (1 + (D / D0)**(2 * n))


# Output Butterworth
SY_butterworth = SI_shift*H_butterworth

Y_butterworth = np.fft.ifft2(SY_butterworth)

Y_butterworth = np.abs(Y_butterworth).astype(np.uint8)

SY_magnitude_butterworth = np.abs(SY_butterworth)
SY_log_magnitude_butterworth = np.log2(SY_magnitude_butterworth)


plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(I, cmap="gray")
plt.title('Original image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(SI_log_magnitude, cmap="gray")
plt.title('Original image spectrum')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(H_butterworth, cmap="gray")
plt.title('Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(SY_log_magnitude_butterworth, cmap="gray")
plt.title('After filter image spectrum')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(Y_butterworth, cmap="gray")
plt.title('After filter')
plt.axis('off')

plt.tight_layout()
```

### Find output with Gaussian LPF

```python
H_gaussian = np.zeros_like(I, dtype=np.float64)

for i in range(h):
    for j in range(w):
        D = np.sqrt((i - center_y)**2 + (j - center_x)**2)
        H_gaussian[i, j]  = np.exp(-(D**2) / (2 * D0**2))

# Output Gaussian
SY_gaussian = SI_shift*H_gaussian

Y_gaussian = np.fft.ifft2(SY_gaussian)

Y_gaussian = np.abs(Y_gaussian).astype(np.uint8)

SY_magnitude_gaussian = np.abs(SY_gaussian)
SY_log_magnitude_gaussian = np.log2(SY_magnitude_gaussian)


plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1)
plt.imshow(I, cmap="gray")
plt.title('Original image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(SI_log_magnitude, cmap="gray")
plt.title('Original image spectrum')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(H_gaussian, cmap="gray")
plt.title('Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(SY_log_magnitude_gaussian, cmap="gray")
plt.title('After filter image spectrum')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(Y_gaussian, cmap="gray")
plt.title('After filter')
plt.axis('off')

plt.tight_layout()
```

```python
F1 = np.array([[1, 2, 1],
               [2, 4, 2],
               [1, 2, 1]]) /4

F2 = np.array([[0, 1, 0],
               [1, 4, 1],
               [0, 1, 0]]) /4

F3 = np.array([[1, 1, 1],
               [1, -8, 1],
               [1, 1, 1]])

F4 = np.array([[1, -2, 1],
               [1, -2, 1],
               [1, -2, 1]])

N = 256
F1 = np.pad(F1, ((0, N-3), (0, N-3)), mode='constant')
F2 = np.pad(F2, ((0, N-3), (0, N-3)), mode='constant')
F3 = np.pad(F3, ((0, N-3), (0, N-3)), mode='constant')
F4 = np.pad(F4, ((0, N-3), (0, N-3)), mode='constant')


F1_fft = np.fft.fft2(F1)
F1_fft_shift = np.fft.fftshift(F1_fft)
magnitude_F1 = np.abs(F1_fft_shift)

F2_fft = np.fft.fft2(F2)
F2_fft_shift = np.fft.fftshift(F2_fft)
magnitude_F2 = np.abs(F2_fft_shift)

F3_fft = np.fft.fft2(F3)
F3_fft_shift = np.fft.fftshift(F3_fft)
magnitude_F3 = np.abs(F3_fft_shift)

F4_fft = np.fft.fft2(F4)
F4_fft_shift = np.fft.fftshift(F4_fft)
magnitude_F4 = np.abs(F4_fft_shift)

plt.figure(figsize=(10, 8))
plt.subplot (2, 2, 1)
plt.imshow(magnitude_F1)
plt.title("F1")
plt.axis("off")

plt.subplot (2, 2, 2)
plt.imshow(magnitude_F2)
plt.title("F2")
plt.axis("off")

plt.subplot (2, 2, 3)
plt.imshow(magnitude_F3)
plt.title("F3")
plt.axis("off")

plt.subplot (2, 2, 4)
plt.imshow(magnitude_F4)
plt.title("F4")
plt.axis("off")
```
