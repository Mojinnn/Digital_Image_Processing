```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Piecewise Linear Transformation Example in book

### Read book & convert to normalize

```python
image_Path = r'D:\py\Digital_Image_Processing\Image\XLA.jpg'

I = cv2.imread(image_Path)
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
I = I.astype(np.uint8)
```

### Locate

```python
L = 256
r1 = (L/4 + L/2) / 2
r2 = (L/2 + 3*L/4) / 2
s1 = (L/4 + 0) / 2
s2 = (L-1 + 3*L/4) / 2
```

### Create output array

```python
Y = np.zeros_like(I)
```

### First piecewise

```python
a, b, p, q = 0, r1, 0, s1 
idx = (I >= a) & (I < b)
Y[idx] = ((I[idx] - a)/(b - a)) * (q - p) + p
```

### Second piecewise

```python
a, b, p, q = r1, r2, s1, s2
idx = (I >= a) & (I < b)
Y[idx] = ((I[idx] - a)/(b - a)) * (q - p) + p
```

### Third piecewise

```python
a, b, p, q = r2, L - 1, s2, L - 1
idx = (I >= a) & (I < b)
Y[idx] = ((I[idx] - a)/(b - a)) * (q - p) + p
```

### Thresholding 

```python
Z = Y.copy()
Z[Z <= 4*L/5] = 0
Z[Z > 0] = 255
```

### Display

```python
plt.figure(figsize= (10, 8))
plt.subplot(1, 3, 1)
plt.imshow(I, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(Y, cmap='gray')
plt.title('PieceWise Linear Transformation Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(Z, cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')
```

```python
# Toa do cau b
r0b = 0
r1b = ((L/5)+2*(L/5))/2
r2b = (2*(L/5) + 3*(L/5))/2
r3b = L - 1
s0b = 0
s1b = ((L/5)+2*(L/5))/2
s2b = 4*(L/5)
s3b = (2*(L/5) + 3*(L/5))/2
s4b = L - 1
```

#### Figure 3.11a

```python
# First piecewise
a, b, p, q = r0a, r1a, s0a, s0a
idx = (I >= a) & (I < b) 
Z[idx] = ((I[idx] - a)/(b - a)) * (q - p) + p

# Second piecewise
a, b, p, q = r1a, r2a, s1a, s1a  
idx = (I >= a) & (I < b) 
Z[idx] = ((I[idx] - a)/(b - a)) * (q - p) + p

# Third piecewise
a, b, p, q = r2a, r3a, s0a, s0a
idx = (I >= a) & (I < b) 
Z[idx] = ((I[idx] - a)/(b - a)) * (q - p) + p
```

#### Figure 3.11b
