```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

# Bai 1: Given a grayscale image I, mask G = [[1, 2, 1], [2, 4, 2], [1, 2, 1]]

```python
image_Path = r'D:\py\Digital_Image_Processing\Image\XLA.jpg'
I = cv2.imread(image_Path, cv2.IMREAD_GRAYSCALE)
G = np.array ([[1, 2, 1],
               [2, 4, 2],
               [1, 2, 1]]) / 16
```

### Cau a: Compute and show Y = I*G

```python
Y = convolve2d(I, G, mode= 'same', boundary= 'symm')
```

```python
plt.figure(figsize=(10,5))
plt.imshow(Y, cmap='gray')
plt.title('Y = I*G')
plt.axis('off')
```

### Cau b: Repeat with G = ones(7, 7)/16, then with G - ones(11, 11)/121

```python
G1 = np.ones((7, 7)) / 49
G2 = np.ones((11, 11)) / 121
Y1 = convolve2d(I, G1, 'same', 'symm' )
Y2 = convolve2d(I, G2, 'same', 'symm' )
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(Y1, cmap='gray')
plt.title('Y1 = I*G1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Y2, cmap='gray')
plt.title('Y2 = I*G1')
plt.axis('off')
```

### Cau d: Add Gaussian noise to I, and reapply the filter G (of various sizes and types)

```python
sigma = 5
noisy_I = I + sigma*np.random.randn(*I.shape)
Y = convolve2d(noisy_I, G, 'same', 'symm')
Y1 = convolve2d(noisy_I, G1, 'same', 'symm')
Y2 = convolve2d(noisy_I, G2, 'same', 'symm')
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.imshow(noisy_I, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(Y, cmap='gray')
plt.title('Y = I_noisy*G')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(Y1, cmap='gray')
plt.title('Y1 = I_noisy*G1')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(Y2, cmap='gray')
plt.title('Y2 = I_noisy*G1')
plt.axis('off')
```

# Bai 2: Give a mask: G = [0 1 0, 1 -4 1, 0 1 0]. Use this mask G to filter a grayscale image I. Oberserve the result and comment

```python
image_Path_2 = r'D:\py\Digital_Image_Processing\Image\hinhlaco.jpg'
I_2 = cv2.imread(image_Path_2, cv2.IMREAD_GRAYSCALE)
G = np.array ([[0, 1, 0],
               [1, -4, 1],
                [0, 1, 0]])
Y = cv2.filter2D(I_2, -1, G)
Y_1 = convolve2d(I_2, G, 'same', 'symm' )
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(I_2, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(Y, cmap='gray')
plt.title('Y = I_2*G')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(Y_1, cmap='gray')
plt.title('Y_1 = I_2*G')
plt.axis('off')
```
