# Homework 1: Create a flag with green background and yellow plus

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Create a rectangle green background

```python
# Image size
height, width = 1000, 1500

# Create an empty image (black) and set it to green
green_image = np.zeros((height, width, 3), dtype=np.uint8)
green_image[:, :, 1] = 255
```

```python
# Display the Green Image
plt.figure()
plt.imshow(green_image)
plt.title("Green Image")
plt.axis("off")
```

### Create a horizontal yellow bar

```python
# Center of the image
center_y, center_x = height // 2, width // 2
# Thickness of the cross
thickness = 100
# Margin from the edges
border = 200
```

```python
# **Horizontal Bar of the Plus (+)**
green_image[center_y - thickness:center_y + thickness, border:width-border, 0] = 255
green_image[center_y - thickness:center_y + thickness, border:width-border, 1] = 255
green_image[center_y - thickness:center_y + thickness, border:width-border, 2] = 0
```

```python
# Display the first part of the cross
plt.figure()
plt.imshow(green_image)
plt.title("Green image with Horizontal bar of the Plus")
plt.axis("off")
```

### Create a vertical yellow bar

```python
# **Vertical Bar of the Plus (+)**
green_image[border:height - border, center_x - thickness:center_x + thickness, 0] = 255
green_image[border:height - border, center_x - thickness:center_x + thickness, 1] = 255
green_image[border:height - border, center_x - thickness:center_x + thickness, 2] = 0
```

```python
plt.figure()
plt.imshow(green_image)
plt.title("Green image with final of the Plus")
plt.axis("off")
```

### Display the size of image

```python
print(f"Image Size : {green_image.shape}")
```

### Define grid size & image file

```python
numRows = 4
numCols = 4
h, w = 400, 400
```

### List of image filenames

```python
image_files = [
    image1_Path, image2_Path, image3_Path, image4_Path,
    image5_Path, image6_Path, image7_Path, image8_Path,
    image9_Path, image1_Path, image11_Path, image12_Path,
    image13_Path, image14_Path, image15_Path, image16_Path
]
```

### Load and resize images

```python
images = []
for img_file in image_files:
    img = cv2.imread(img_file)
    if img is None:
        print(f"Cannot read {img_file}")
        continue
    img = cv2.resize(img, (h, w))
    images.append(img)
```

### Create a blank big image & place each image in the correct position

```python
big_Image = np.zeros((numRows*h, numCols*w, 3), dtype=np.uint8)

for i in range(numRows):
    for j in range(numCols):
        index = i*numCols + j
        x_start, x_end = j*w, (j+1)*w
        y_start, y_end = i*h, (i+1)*h
    
        big_Image[y_start:y_end, x_start:x_end] = images[index]
```

### Convert BGR to RGB for correct display & display

```python
big_Image_rgb = cv2.cvtColor(big_Image, cv2.COLOR_BGR2RGB)

plt.imshow(big_Image_rgb)
plt.title("Merged Image")
plt.axis("off")
plt.show()
```
