```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = r'D:\py\Digital_Image_Processing\Image\rectangle.jpg'
I = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
kernel_a = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

kernel_b = np.ones((5, 5), np.uint8) 

I_erosion = cv2.erode(I, kernel_b, iterations=10)
I_dilation = cv2.dilate(I, kernel_b, iterations=10)

plt.figure(figsize=(10, 12))
plt.subplot(1, 3, 1)
plt.imshow(I, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(I_erosion, cmap='gray')
plt.title('Erosion Image')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.imshow(I_dilation, cmap='gray')
plt.title('Dilation Image')
plt.axis('off')
```

```python
cv2opening = cv2.morphologyEx(I, cv2.MORPH_OPEN, kernel_b, iterations=10)
cv2closing = cv2.morphologyEx(I, cv2.MORPH_CLOSE, kernel_b, iterations=10)


plt.figure(figsize=(10, 12))
plt.subplot(1, 3, 1)
plt.imshow(I, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2opening, cmap='gray')
plt.title('Openning Image')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.imshow(cv2closing, cmap='gray')
plt.title('Closing Image')
plt.axis('off')
```

## Fill in the space inside the disks without changing the exterior boundaries of the disk

```python
path =  r'D:\py\Digital_Image_Processing\Image\circles.png'
I2 = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
```

```python
I2_dilation = cv2.dilate(I2, kernel_a, iterations=4)
I2_erosion = cv2.erode(I2_dilation, kernel_a, iterations=4)
I2_erosion = cv2.erode(I2_erosion, kernel_a, iterations=4)


result = I2_erosion | I2

plt.figure(figsize=(10, 12))
plt.subplot(1, 3, 1)
plt.imshow(I2, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(I2_erosion, cmap='gray')
plt.title(' Image')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.imshow(result, cmap='gray')
plt.title('Result Image')
plt.axis('off')
```
