# Demo exp & log function 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read image & normalization

```python
image_Path = r'D:\py\Digital_Image_Processing\Image\XLA.jpg'

I = cv2.imread(image_Path)
I_gray = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
I_gray = I_gray.astype(np.float64) / 255 # Chuan hoa anh tu [0:1]
```

### Function implementation

```python
gamma1 = 0.3
Y1 = np.power(I_gray, gamma1)

gamma2 = 3
Y2 = np.power(I_gray, gamma2)

Y3 = np.log(1 + I_gray)
```

### Normalize to display with plt

```python
Y1 = np.uint8(Y1 * 255)
Y2 = np.uint8(Y2 * 255)
Y3 = np.uint8(Y3 / np.max(Y3) * 255)
```

### Display

```python
plt.figure(figsize= (10, 7))

plt.subplot(2, 2, 1)
plt.imshow(I_gray, cmap = 'gray')
plt.title(f'Original \nSize = {I_gray.shape}')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(Y1, cmap='gray')
plt.title(f'gamma = 0.3 \nSize = {Y1.shape}')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(Y2, cmap='gray')
plt.title(f'gamma = 3 \nSize = {Y2.shape}')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(Y3, cmap='gray')
plt.title(f'log \nSize = {Y3.shape}')
plt.axis('off')
```
